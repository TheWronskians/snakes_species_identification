```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

from scipy import misc
import sys
import os
import argparse
import tensorflow as tf
import numpy as np
import random
from PIL import Image
import argparse
from time import sleep
import numpy as np
import imageio
import glob
import matplotlib.pyplot as plt
import cv2
```

```python
def readImage(imagepath):
    im =  None
    try:
        image = Image.open(imagepath)
        image.copy().verify()
        im = np.asarray(image)
        im = cv2.cvtColor(im,cv2.COLOR_RGB2BGR)
    except (IOError,SyntaxError) as e:
        print('Bad file:', imagepath)
    return im
```

```python
old_path = "train/"
for name in os.listdir(old_path):
    a = []
    for file in os.listdir(old_path + "/" + name):
        a.append(file)
    new_path_train = "datasets/train/" + name
    new_path_val = "datasets/val/" + name
    os.mkdir(new_path_train)
    os.mkdir(new_path_val)
    for i in range(len(a)):
        if i % 5 == 0:
            curr_img = readImage(old_path + name + "/"  + a[i])
            if curr_img is not None:
                cv2.imwrite(new_path_val + "/" + a[i],curr_img)
        else:
            curr_img = readImage(old_path + name + "/"  + a[i])
            if curr_img is not None:
                cv2.imwrite(new_path_train + "/" + a[i],curr_img)
```
